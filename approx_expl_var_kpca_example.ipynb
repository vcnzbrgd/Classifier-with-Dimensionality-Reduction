```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import KernelPCA
from sklearn.datasets import make_circles

# Generate a sample dataset
X, _ = make_circles(n_samples=400, factor=.3, noise=.05)

# Fit Kernel PCA
kpca = KernelPCA(kernel="rbf", fit_inverse_transform=True, gamma=10)
X_kpca = kpca.fit_transform(X)

# Compute normalized eigenvalues as proxy for explained variance
eigenvalues = np.sort(kpca.eigenvalues_)[::-1]  # Sort eigenvalues in descending order
explained_variance_ratio = eigenvalues / eigenvalues.sum()

# Plot the explained variance
plt.figure(figsize=(8, 4))
plt.plot(np.cumsum(explained_variance_ratio), marker='o')
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.title('Approximated Explained Variance by Kernel PCA Components')
plt.grid(True)
plt.show()
```
